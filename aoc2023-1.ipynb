```python
elf_raw = file_contents
```

```python
elf_list = elf_raw.split("\n")
```

```python
full_elf_sum = 0
for i in elf_list[0:-1]:
    real_calibration = str()
    for k in i:
        if k.isdigit() == True:
            real_calibration += k
    
    final_calibration = real_calibration[0] + real_calibration[-1]
    print(final_calibration)
    full_elf_sum = full_elf_sum + int(final_calibration)
```
